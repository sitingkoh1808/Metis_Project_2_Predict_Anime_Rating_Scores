```python
import re

def get_info(url):
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page,"lxml")
    
    headers = ['title','ranking','users_votes','score','popularity', 
               'favorites','type','episode','year','premiered','members','source','studios']
               
    info = soup.find_all('div', class_="spaceit")
    
    #get title
    title_string = soup.find('h1', class_='title-name')
    title=title_string.text
    
    #get episodes 
    info_epi = info[0].text
    episodes=re.sub("[^0-9]", "", info_epi)
    
    #get year
    info_year = info[1].text
    year=re.sub("[^0-9]", "", info_year[17:22])
    
    #get members
    info_members = soup.find('span', class_="numbers members")
    members = re.sub("[^0-9]", "",info_members.find('strong').text)

    #get score 
    score = soup.find('div', class_="fl-l score").text
    
    #get rank 
    info_rank = soup.find('span', class_="numbers ranked")
    rank = re.sub("[^0-9]", "", info_rank.text)
    
    #get users_votes
    info_user = soup.find('div', class_="fl-l score")
    user = re.sub("[^0-9]", "", info_user.get('data-user'))
    
    #get popularity
    info_popularity = soup.find('span', class_="numbers popularity")
    popularity = re.sub("[^0-9]", "", info_popularity.text)
    
    #get favorites
    fav = soup.find('table')
    fav_rows = [row for row in fav.find_all('div')]
    favorites=''
    for i in fav_rows[24:]:
        a = i.text
        b = a[0:11].strip()
        if b == 'Favorites:':
            favorites=re.sub("[^0-9]", "", i.text)
            break
    
   #get type
    types=''
    for i in fav_rows[8:12]:
        c = i.text
        d = c[0:6].strip()
        if d == 'Type:':
            types = c.split(':')[1].strip()
            break
    
    #get source       
    source=''
    for i in fav_rows[15:22]:
        e = i.text
        f = e[0:8].strip()
        if f == 'Source:':
            source = e.split(':')[1].strip()
            break
    
    #get studios
    studios=''
    for i in fav_rows[15:20]:
        g = i.text
        h = g[0:9].strip()
        if h == 'Studios:':
            studios = g.split(':')[1].strip()
            break
    
#     #get rating 
#     rating=''
#     for i in fav_rows[20:25]:
#         j = i.text
#         k = j[0:8].strip()
#         if k == 'Rating:':
#             rating = j.split(':')[1].strip().split('-')[1]
#             break

    #get premiered 
    premiered =''
    for i in fav_rows[10:16]:
        l = i.text
        m = l[0:11].strip()
        if m == 'Premiered:':
            premiered = l.split(':')[1].split(' ')[0].strip()
            break
    
    ##Create movie dictionary and return
    anime_dict = dict(zip(headers, [title,
                                    rank,
                                    user,
                                    score,
                                    popularity,
                                    favorites,
                                    types,
                                    episodes,
                                    year,
                                    premiered,
                                    members,
                                    source,
                                    studios]))

    return anime_dict
```

```python
animes_info_list = []

for url in animes_df2.url:
    animes_info_list.append(get_info(url))
    sleep(2)
```

```python
animes_info_df = pd.DataFrame(animes_info_list)  #convert list of dict to df
# animes_info_df.set_index('title', inplace=True)

animes_info_df
```
