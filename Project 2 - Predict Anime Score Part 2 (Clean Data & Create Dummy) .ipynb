```python
type_source_df = type_df.merge(source_df, left_on=['users_votes','score','popularity','favorites','episode','year',
                                                  'members'], right_on=['users_votes','score','popularity',
                                                    'favorites','episode','year','members'])
type_source_df
```

```python
type_source_studios_df=type_source_df.merge(studios_df, left_on=['users_votes','score','popularity','favorites','episode','year',
                                                  'members'], right_on=['users_votes','score','popularity',
                                                    'favorites','episode','year','members'])
```

```python
mal_df=type_source_studios_df.merge(premiered_df, left_on=['users_votes','score','popularity','favorites','episode','year',
                                                  'members'], right_on=['users_votes','score','popularity',
                                                    'favorites','episode','year','members'])
mal_df
```
